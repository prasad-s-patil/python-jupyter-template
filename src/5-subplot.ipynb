### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, in the fourth.
```

```python
import matplotlib.pyplot as plt

# Data for the plots
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot: Line chart
axs[0, 0].plot(x, y, marker='o')
axs[0, 0].set_title('Line Chart')

# Second subplot: Bar chart
axs[0, 1].bar(x, y)
axs[0, 1].set_title('Bar Chart')

# Third subplot: Scatter plot
axs[1, 0].scatter(x, y)
axs[1, 0].set_title('Scatter Plot')

# Fourth subplot: Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%')
axs[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
