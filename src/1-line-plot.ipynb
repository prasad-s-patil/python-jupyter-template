### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python

import matplotlib.pyplot as plt
import numpy as np

# Generate data: x values ranging from 0 to 10 and y values as the square of x
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = x**2  # y is the square of x

# Create the plot
plt.plot(x, y)

# Customize the plot
plt.title('Line Plot of x vs. x^2')  # Title
plt.xlabel('x')  # x-axis label
plt.ylabel('x^2')  # y-axis label
plt.grid(True)  # Add grid

# Display the plot
plt.show()
```
